```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load csv file
df = pd.read_csv('data2/users.csv')
#df
```

### User "age"

```python
# Simple "age" plot
df["age"].value_counts().plot.pie()
plt.show()
```

```python
# Fill NaNs with -1 and plot "age"
df["age"].fillna(-1).value_counts().plot.pie()
plt.show()
```

### User "gender"

```python
df["gender"].fillna(-1).value_counts().plot.bar()
plt.show()
```

### User "body_type"

```python
df["body_type"].fillna(-1).value_counts().plot.bar()
plt.show()
```

### Location

```python
df["country"].fillna(-1).value_counts().plot.pie()
plt.show()
df["country"].value_counts().plot.pie()
plt.show()
```
