```python
import gym
env = gym.make('Copy-v0')

# Run few episodes.
for ep in range(5):
    # Start new episode
    print("! New episode !")
    # Start new episode and get initial state.
    obs = env.reset()
    env.render()

    # Pefson n steps until the end of sequence.
    for step in range(10):
        print("Executing ",str(step),"-th action")
        # Next action: move right, write to what has been reed to output.
        act= (1, 1, obs)
        print("act=",act)
        # Environment returns observation, reward, and boolean flag indicating if the episode is complete.
        obs,rew,done,_ = env.step(act) 
        print("step=",step)
        print("obs=",obs)
        #print("rew=",rew)
        print("done=",done)
        env.render()
        if done == True:
            break
    
```
